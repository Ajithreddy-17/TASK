```python


import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('hospital data analysis.csv')

# Average cost by age
data = df.groupby('Age')['Cost'].mean()

plt.figure(figsize=(10,5))
plt.plot(data.index, data.values, marker='o')
plt.title('Average Cost by Age')
plt.xlabel('Age')
plt.ylabel('Average Cost')
plt.grid(True)
plt.show()


# Average cost by condition
data = df.groupby('Condition')['Cost'].mean()

plt.figure(figsize=(8,5))
plt.bar(data.index, data.values)
plt.title('Average Cost by Condition')
plt.xlabel('Condition')
plt.ylabel('Average Cost')
plt.xticks(rotation=45)
plt.show()


data = df.groupby('Condition')['Cost'].mean()

plt.figure(figsize=(10,5))
plt.scatter(data.index, data.values)
plt.title('Average Cost by Condition')
plt.xlabel('Condition')
plt.ylabel('Average Cost')
plt.xticks(rotation=45)
plt.show()

```
