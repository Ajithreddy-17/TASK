```python

import pandas as pd
orders = pd.read_csv('olist_orders_dataset.csv')
orders.head()


orders['order_purchase_timestamp'] = pd.to_datetime(
    orders['order_purchase_timestamp']
)


orders['month'] = orders['order_purchase_timestamp'].dt.strftime('%y-%m')


monthly_orders = orders.groupby('month')['order_id'].count()


print("Month with Highest Orders:")
print(monthly_orders.idxmax())
print("Number of Orders:", monthly_orders.max())

import matplotlib.pyplot as plt


monthly_orders = orders.groupby('month')['order_id'].count()

# Plot
monthly_orders.plot(kind='bar', figsize=(12,5))
plt.title('Number of Orders by Month')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()

```

```python
import pandas as pd


order_items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
translation = pd.read_csv('product_category_name_translation.csv')


df = pd.merge(order_items, products, on='product_id')


df = pd.merge(df, translation, on='product_category_name')


top_categories = df.groupby('product_category_name_english')['order_id'].count()


top_categories = top_categories.sort_values(ascending=False)


print("Top 10 Product Categories by Number of Orders:")
print(top_categories.head(10))


import matplotlib.pyplot as plt

# Top 10 categories by orders
top_categories = (
    df.groupby('product_category_name_english')['order_id']
      .count()
      .sort_values(ascending=False)
      .head(10)
)

# Plot
top_categories.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Product Categories by Number of Orders')
plt.xlabel('Product Category')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd


order_items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
translation = pd.read_csv('product_category_name_translation.csv')

df = pd.merge(order_items, products, on='product_id')
df = pd.merge(df, translation, on='product_category_name')


category_revenue = df.groupby('product_category_name_english')['price'].sum()


category_revenue = category_revenue.sort_values(ascending=False)


print("Top 10 Product Categories by Revenue:")
print(category_revenue.head(10))


import matplotlib.pyplot as plt

# Top 10 categories by revenue
category_revenue = (
    df.groupby('product_category_name_english')['price']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

# Plot
category_revenue.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Product Categories by Revenue')
plt.xlabel('Product Category')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```
