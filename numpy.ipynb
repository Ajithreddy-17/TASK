```python
import pandas as pd
s=pd.Series([10,20,30,40,50,60])
print(s.describe())
print(s.mean())
print(s.sum())
print(s.min())
print(s.max())

```

```python
s1=pd.Series([11,12,13,14,15])
print(s1.map(lambda x : x**2))
s2=pd.Series([29,27,11,27,49,1,10000])
print(s2.sort_values())
print(s2.drop(2))
print(s2.duplicated())
```

```python
####handle missing values
import numpy as np
s3=pd.Series([10,np.nan,18,19,np.nan,28])
print(s3.isnull())
print(s3.notnull())
print(s3.fillna(22))
print(s3.dropna())
```

```python
import numpy as np
import pandas as pd
s4=pd.Series([19,20,21,22,23,24,25])
print(s4.iloc[1])
print(s4.iloc[-2])
print(s4.loc[3])
print(s4[s4>=21])
```
